```python
import cv2
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("Tulip.jpg")
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(image= img, threshold1 =127, threshold2 = 127)
```

```python
plt.imshow(edges)
```

```python
med_values= np.median(img)
```

```python
med_values
```

```python
lower= int(max(0,0.7*med_values))
upper=int(min(255,1.3*med_values))

edges =cv2.Canny(img, threshold1=lower, threshold2=upper)

plt.imshow(edges)
```

```python
blurred_img=cv2.blur(img, ksize=(5,5))

edges= cv2.Canny(image=blurred_img,
                threshold1=lower,
                threshold2=upper)

plt.imshow(edges)
```
